```python
from PIL import Image
import os

# Define the path to the dataset and the target size for the images
dataset_path = "D:\\NSBM\\lecture materials\\3rd year\\skindiseases\\training\\Dataset"
target_width, target_height = 224, 224  # Define the target size for the images

# Function to resize an image to the target size
def resize_image(image_path, target_size):
    with Image.open(image_path) as img:  # Use 'with' for context management
        img_resized = img.resize(target_size, Image.LANCZOS)
        return img_resized

# Function to process each folder in the dataset
def process_folder(folder_path, target_size):
    # Get the list of image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]
    
    # Create an output folder for resized images
    output_folder = os.path.join(folder_path, "resized")
    os.makedirs(output_folder, exist_ok=True)
    
    # Iterate through each image file
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        
        # Resize the image
        resized_image = resize_image(image_path, target_size)
        
        # Save the resized image in the output folder
        resized_image_path = os.path.join(output_folder, image_file)
        resized_image.save(resized_image_path)
        
        print(f"Resized and saved image: {resized_image_path}")

# Define the target size for the images
target_size = (target_width, target_height)

# Iterate through each folder in the dataset
folders = ["Eczema", "Melanoma", "Atopic Dermatitis", "Melanocytic Nevi (NV)", "Warts Molluscum and other Viral Infections", "BA-impetigo", "Gianotti-Crosti syndrome", "Rashes in systemic diseases", "Vascular malformation", "VI-chickenpox"]

for folder_name in folders:
    folder_path = os.path.join(dataset_path, folder_name)
    if os.path.isdir(folder_path):
        process_folder(folder_path, target_size)
    else:
        print(f"Folder not found: {folder_path}")

print("Image resizing complete.")
```
